```
import  matplotlib.pyplot as plt
import numpy as np
```

```
# for jupyter notebook for displaying output 
%matplotlib inline
```

```
x = np.arange(1,20)
y = x ** 2

plt.plot(x,y)
plt.xlabel('Domaain')
plt.ylabel('Range')
plt.title('Simple Graph')
plt.show()
```

### <p style="text-align:center">Drawing Multiple Plot in Same Canvas Using plt.figure() Object</p>

```
# subplot (number_of_row, number_of_column,plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'red')

plt.subplot(1,2,2)
plt.plot(x,y,'blue')
```

### <p style="text-align:center"> Creating Empty Canvas and Ploting more graphs inside it</p>

```
# Creating an Empty Canvas
figure  = plt.figure()

# figure.add_axes([left,bottom,width,height])
outer_axes = figure.add_axes([0.1,0.1,0.8,0.8])
outer_axes.plot(x,y)
outer_axes.set_title('Outer Plot')

inner_axes = figure.add_axes([0.2,0.5,0.4,0.3])
inner_axes.plot(y,x)
inner_axes.set_title('Inner Plot')
```

### <p style="text-align:center">Drawing Multiple Plot in Same Canvas using plt.subplots() Object</p>

```
fig,axes = plt.subplots(nrows=1,ncols=2)
#fig,axes = plt.subplots(1,2)

for axis in axes :
    axis.plot(x,y)
```

```
# figsize=(width_inches,height_inches)
another_fig , another_axes = plt.subplots(2,2,figsize=(10,10))

for i in range(2) :
    for j in range(2):
        #another_axes[i][j].plot(x,y) does the same job
        another_axes[i,j].plot(x,y)
```

### <p style="text-align:center"> Ploting Two Graph inside a Single Graph</p>

```
new_graph = plt.figure()

new_axis = new_graph.add_axes([0,0,1,1])


# linewidht is lw and linestyle is ls here
new_axis.plot(x,x**2,label='X Squared',color='red',linestyle='--',linewidth='5')
new_axis.plot(x,x**3, label = 'X Cubed',color='green',ls='-.',lw='6',marker='o',markerfacecolor='blue',markersize=10)

new_axis.legend(loc=10)
```
